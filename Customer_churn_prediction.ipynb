```python
# convert string values (yes or no ) of churn column to 1 and 0
df.loc[df.Churn=='No','Churn'] = 0
df.loc[df.Churn=='Yes','Churn'] = 1
```

```python
# convert 'No internet service' to 'No' for the columns
cols=['OnlineBackup','StreamingMovies','DeviceProtection','TechSupport','OnlineSecurity','StreamingTV']
for i in cols:
    df[i]=df[i].replace({'No internet service':'No'})
```

```python
df.head()
```

```python
df['MultipleLines']=df['MultipleLines'].replace({'No phone service':'No'})
df.head()
```

```python
import numpy as np
df['TotalCharges'] = df['TotalCharges'].replace(" ",np.nan)
```

```python
# drop null values of 'TotalCharges' feature
df=df[df['TotalCharges'].notnull()]
df=df.reset_index()[df.columns]
```

```python
# convert 'TotalCharges' column values to float datatype
df['TotalCharges']=df['TotalCharges'].astype(float)
```

```python
# training with hyperparameters
models={
    "Decision Tree":DecisionTreeClassifier(random_state=42),
    "Random Forest":RandomForestClassifier(random_state=42)
}
```

```python
cv_scores={}
for model_name,model in models.items():
    scores=cross_val_score(model,X_train,y_train,cv=5,scoring='accuracy')
    cv_scores[model_name]=scores
    print(f"{model_name} cross validation accuracy: {np.mean(scores):.2f}")
```
